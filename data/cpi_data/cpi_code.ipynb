Above are line plots for each of the categories from the CPI that we utilized as part of our data set.

When we discuss CPI values, we are not listing an exact price that certain items are bought/sold. Rather, a CPI value is measuring is a relative change in prices for the given category of goods.

For example, if data showed that in January 2021, toys had a CPI value of 120, and then in February of 2021 they had CPI value of 190 then we would know that the average price of toys saw an increase over time between January and February. 

However, in addition to looking at the overall trend of these line plots, and seeing that a significant proportion of these goods experienced an increase in their average price, we can calculate the overall inflation percent change for each categorty to fully understand these price changes. 

In order to calculate the overall infaltion percent change, we are using the formula:

$$
Inflation = \frac {End time - Start time}{Start time} * 100
$$

```python
# Code to calculate the overall inflation of each category of goods included from CPI data

# Ensure numeric CPI columns
all_cpi.iloc[:, 2:] = all_cpi.iloc[:, 2:].apply(pd.to_numeric, errors='coerce')

# Filter for January 2021 and January 2025
start_values = all_cpi[(all_cpi["Year"] == 2021) & (all_cpi["Month"] == "Jan")].iloc[:, 2:]
end_values = all_cpi[(all_cpi["Year"] == 2025) & (all_cpi["Month"] == "Jan")].iloc[:, 2:]

# Calculate overall inflation
overall_inflation = ((end_values.values - start_values.values) / start_values.values) * 100

# Convert to DataFrame
overall_inflation_df = pd.DataFrame(overall_inflation, columns=start_values.columns, index=["Overall_Inflation"])

overall_inflation_df
```

```python
# Reset index so we can use category names as x-axis labels
overall_inflation_df = overall_inflation_df.T.reset_index()
overall_inflation_df.columns = ["Category", "Overall Inflation (%)"]

# Sort values for better visualization
overall_inflation_df = overall_inflation_df.sort_values(by="Overall Inflation (%)", ascending=False)

# Plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=overall_inflation_df, x="Overall Inflation (%)", y="Category", palette="coolwarm")

# Add labels and title
plt.xlabel("Inflation Rate (%)")
plt.ylabel("Category")
plt.title("Overall Inflation (Jan 2021 - Jan 2025) by Category")

# Show the plot
plt.show()
```
